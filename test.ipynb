```python
import pandas as pd
```

```python
df= pd.read_csv('HackUTD-2023-HomeBuyerInfo.csv')
df["LTV"]= (df["LoanAmount"] - df["DownPayment"])/df["AppraisedValue"]
df["DTI"]= (df["CarPayment"]+ df["CreditCardPayment"] + df["StudentLoanPayments"] + df["MonthlyMortgagePayment"])/df["GrossMonthlyIncome"]
df["FEDTI"]= df["MonthlyMortgagePayment"]/df["GrossMonthlyIncome"]
df
```

```python
df.columns
```

```python
def rules(x):
    if x["CreditScore"] >=640:
        if x["LTV"]<.80:
            if x["DTI"]<.46:
                if x["FEDTI"]<.28:
                    return "Approved"
    return "Denied"

df["Status"]= df.apply(rules, axis=1)
df
```
