```python
import pandas as pd
```

```python
df= pd.read_csv('HackUTD-2023-HomeBuyerInfo.csv')
df["LTV"]= (df["LoanAmount"] - df["DownPayment"])/df["AppraisedValue"]
df["DTI"]= (df["CarPayment"]+ df["CreditCardPayment"] + df["StudentLoanPayments"] + df["MonthlyMortgagePayment"])/df["GrossMonthlyIncome"]
df["FEDTI"]= df["MonthlyMortgagePayment"]/df["GrossMonthlyIncome"]
df
```

```python
import plotly.express as px
```

```python
df["Approved CreditScore"]= (df["CreditScore"] >=640)*1
df["Approved LTV"]= (df["LTV"]<.80)*1
df["Approved DTI"]= (df["DTI"]<.46)*1
df["Approved FEDTI"]= (df["FEDTI"]<.28)*1
df
```

```python
df["Approved"]= (df["Approved CreditScore"] + df["Approved LTV"] + df["Approved DTI"] + df["Approved FEDTI"] == 4)*1
#Change the ones into positive text and zeros into negative text
df["Approved"]= df["Approved"].replace({1: "Approved", 0: "Not Approved"})
df["Approved CreditScore"]= df["Approved CreditScore"].replace({1: "Approved by Credit Score", 0: "Not Approved by Credit Score"})
df["Approved LTV"]= df["Approved LTV"].replace({1: "Approved by LTV", 0: "Not Approved by LTV"})
df["Approved DTI"]= df["Approved DTI"].replace({1: "Approved by DTI", 0: "Not Approved by DTI"})
df["Approved FEDTI"]= df["Approved FEDTI"].replace({1: "Approved by FEDTI", 0: "Not Approved by FEDTI"})
df
```

```python
fig= px.sunburst(df, path=['Approved CreditScore', 'Approved LTV', 'Approved DTI', 'Approved FEDTI', 'Approved',], color='Approved', color_continuous_scale='RdBu', width=800, height=800,
            title="Approved Home Buyers by Credit Score, LTV, DTI, and FEDTI", color_discrete_map={'Not Approved': 'red', 'Approved': 'green'})
fig.update_layout(margin=dict(t=0, l=0, r=0, b=0))
fig.update_traces(textinfo='label+percent parent')
fig.show()
```

```python
import requests
import json
import tqdm

latitudes= []
longitudes= []

# Replace with your API key
api_key = 'AIzaSyCdfNS1niuPbNhZ7P3mvTIoZu9D-RWs6ZU'

def geocode_place(place):
    base_url = 'https://maps.googleapis.com/maps/api/geocode/json'
    params = {
        'address': place,
        'key': api_key,
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        if data['status'] == 'OK':
            location = data['results'][0]['geometry']['location']
            lat = location['lat']
            lng = location['lng']
            return lat, lng
        else:
            print(f"Geocoding failed for {place}: {data['status']}")
    else:
        print(f"Failed to connect to the Geocoding API. Status code: {response.status_code}")

places = df_places["Address"]

for place in tqdm.tqdm(places):
    lat, lng = geocode_place(place)
    latitudes.append(lat)
    longitudes.append(lng)

df_places["Latitude"]= latitudes
df_places["Longitude"]= longitudes
df_places
```
